```python
%matplotlib inline
import pytry
import pandas
```

```python
data = pytry.read('exp1')
```

```python
data2 = []
for d in data:
    error = d['error']
    del d['error']
    del d['average']
    for distance, errors in error.items():
        for e in errors:
            d['distance'] = distance
            d['error'] = e
            data2.append(dict(d))
```

```python
df = pandas.DataFrame(data2)
```

```python
import seaborn
```

```python
seaborn.barplot('distance', 'error', data=df, hue='n_neurons')
```
